```python
import pyautogui
import time
#moveTo위치를 정해놓고 그 값으로 정확하게 이동할 때 사용한다.
pyautogui.moveTo(x=1299, y=99)
#moveRel지금 주어진 위치에서 x만 몇 픽셀, y만 몇 픽셀 이동하고 싶으면 사용한다.
pyautogui.moveRel(0,0,2)
#마우스 제어
#더블클릭
pyautogui.doubleClick()
#한번클릭
pyautogui.click(clicks=2,interval=1)
#두번클릭
pyautogui.click(clicks=2,interval=2)

time.sleep(1) #더블클릭하는 시간을 고려하지 않았기 때문에 글자를 작성하기전에 언제까지 기다릴수가 없어서 글자를 못씀 그래서 1초 쉬고 typewrite해결해줌 

pyautogui.typewrite('Hello')
pyautogui.typewrite(['enter'])
pyautogui.typewrite(['a','b','c','enter']) #a,b,c키, enter키
pyautogui.typewrite(['abc','enter']) #abc라는 키가 없기 때문에 메모장에 적어주지 않음
```
